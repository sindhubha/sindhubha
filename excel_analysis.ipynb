```python
new_df = df['frequency'].value_counts()
print(new_df)
```

```python
new_df = df['frequency'].value_counts().rename_axis('sub_value').reset_index(name='counts')
new_df[:10]
```

```python
your_lables = new_df.sub_value
your_values = new_df.counts
```

```python
import matplotlib.pyplot as plt
fig = plt.figure() 
ax =fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(your_values, labels = your_lables, autopct='%1.2f%%')
plt.title("frequency_percentage")
plt.show()
```

```python
percentage = []
for i in range(new_df.shape[0]):
  pct=(new_df.counts[i]/your_lables)*100
  percentage.append(round(pct,2))
```

```python
new_df['Percentage'] = percentage
print(new_df)
```
